```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

n=2
insertion_sort=8*n**2
merge_sort=64*n*np.log2(n)
    #Create array to store all values:
ins_sort_runtime = []
merge_sort_runtime = []

while insertion_sort <= merge_sort:
    insertion_sort=8*n**2
    ins_sort_runtime.append(insertion_sort)
    merge_sort=int(64*n*np.log2(n))
    merge_sort_runtime.append(merge_sort)
    n+=1
print("n <=", n-2)
```

```python
#Create dataframe for nicer output and for use with graph 
z=list(np.arange(2,n))
cols = {"n":z, "Insertion Sort":ins_sort_runtime, "Merge Sort":merge_sort_runtime}
df = pd.DataFrame(cols)
df.set_index("n",inplace=True)
display(df)
df=df.reset_index()
```

```python
#Create graphs from dataframe
ax=plt.gca()
plt.style.use('seaborn')

plt.title("When does implementation sort beat merge sort?", fontsize=16)
df.plot(kind="line", x="n", y="Insertion Sort",ax=ax)
df.plot(kind="line", x="n", y="Merge Sort",color="orange",ax=ax)
plt.xlabel("Value of n",fontsize=14)
plt.ylabel("Steps",fontsize=14)
plt.plot(43,14933,'o',color='red',markersize=10) 
ax.annotate("n <=43",xy=(25,14800),xytext=(38,14900),fontsize=14,color="red")
plt.tight_layout
plt.savefig('ques1.2-2.png',dpi=300,bbox_inches="tight")
plt.legend(loc=2, prop={'size': 10})
plt.show()
```
